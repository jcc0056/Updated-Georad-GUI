# GeoRAD Launcher

Open this notebook and **Run All**.

What it does:
1. Prints which Python environment (kernel) you're using
2. Installs GeoRAD from the local `georad-master/` folder (editable install) **into this kernel**
3. Launches the GeoRAD dashboard

Notes:
- If you want OpenAQ features, you'll need an OpenAQ API key. If you don’t have one set as an environment variable, the OpenAQ panel will prompt you for it.


```python
import sys, pathlib

print("Python:", sys.version.split()[0])
print("Executable:", sys.executable)
print("Notebook folder:", pathlib.Path().resolve())
```

```python
import sys, subprocess, pathlib, importlib

root = pathlib.Path().resolve()
project_dir = root / "georad-master"

if not project_dir.exists():
    raise FileNotFoundError(f"Expected folder not found: {project_dir}")

def run(cmd):
    print("+", " ".join(map(str, cmd)))
    subprocess.check_call(list(map(str, cmd)))

def ensure_georad_installed():
    try:
        import georad              
        import pathlib as _p
        import georad as _georad
        installed_from = _p.Path(_georad.__file__).resolve()
        if project_dir not in installed_from.parents:
            print("GeoRAD is currently imported from:")
            print(" ", installed_from)
            print("Re-pointing editable install to this folder...")
            run([sys.executable, "-m", "pip", "install", "-e", str(project_dir)])
            importlib.invalidate_caches()
            if "georad" in sys.modules:
                del sys.modules["georad"]
    except Exception:
        print("Installing GeoRAD into this kernel (editable mode)...")
        run([sys.executable, "-m", "pip", "install", "-e", str(project_dir)])

ensure_georad_installed()

import georad
print("GeoRAD version:", getattr(georad, "__version__", "unknown"))
print("GeoRAD imported from:", georad.__file__)
```

```python
import os
key = os.getenv("OPENAQ_API_KEY")
if key:
    print("OPENAQ_API_KEY is set (length:", len(key), ")")
else:
    print("OPENAQ_API_KEY is NOT set.")
    print("That's okay—GeoRAD will still start. OpenAQ panels will ask for a key when needed.")
```
