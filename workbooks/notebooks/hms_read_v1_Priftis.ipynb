# Problem Statement


Smoke pollution from wildfires in the western United States has become an increasingly serious public health concern, especially following the record-breaking fire seasons in 2018, 2020, and 2021 (Burke et al., 2021; Zhou et al., 2021). In the recent years fires over the U.S. have increased, with only in California have burnt nearly 7 million acres in the last decade. Decades of fire suppression in the 20th century, combined with droughts and a warming climate, have led to longer and more intense fire seasons, marked by megafires that are challenging to control (Williams et al., 2019; Juang et al., 2022). The growing population in the wildland-urban interface is vulnerable to wildfires and, in turn, may contribute to accidental ignitions. Wildfire smoke pollution has clear impacts on public health, affecting communities along the U.S. West Coast. Utilizing a synergy of satellite systems, NOAA has developed the Mapping Hazard System (Rolph et al., 2009; Brey et al. 2018), which identifies smoke plume regions across the Continental U.S. (CONUS). Case studies can leverage such information to examine the sptiatemporal impact of smoke plumes for specific atmospheric conditions.

This information is essential for emergency responders, environmental agencies, and public health officials as it supports decision-making processes that protect communities and ecosystems from fire-related risks. Additionally, smoke monitoring helps identify air quality impacts over large areas, assisting in the issuance of health advisories and promoting public safety. NOAA’s consistent and accurate monitoring is fundamental to understanding fire behavior, assessing its atmospheric effects, and developing long-term strategies to adapt to increasing wildfire risks driven by climate change.



# Dataset Description

The **Hazard Mapping System (HMS) Fire and Smoke Product** is a tool developed by the United States National Oceanic and Atmospheric Administration (NOAA) to monitor and provide information about wildfires and smoke dispersion.


The goal of the HMS is to provide timely and accurate information to emergency responders, decision-makers, and the public to help them make informed decisions and take appropriate actions in response to these hazards. The HMS Fire and Smoke Product combines satellite data, weather information, and modeling to create maps that show the ***location and intensity of wildfires***, as well as the ***movement and dispersion of smoke plumes*** resulting from these fires.


**What satellite/sensors does HMS use?**


Currently, the HMS uses the 2km resolution ABI data from GOES-16 (East) and GOES-17 (West) geostationary satellites, along with 375m resolution VIIRS data from S-NPP and NOAA-20 polar satellites, and 1km resolution MODIS data from the Terra polar satellite. Complementary USGS 30 m resolution Landsat-7 Enhanced Thematic Mapper Plus (ETM+) and Landsat-8 Operational Land Imager (OLI), and ESA 20 m resolution Sentinel-2A/B MultiSpectral Instrument (MSI) data are also used in support of the daily fire analyses. A key component is the quality control performed by an analyst who inspects all available imagery and automated fire detects, deletes those detects believed to be false alarms and adds additional fires that the automated routines have not detected.



Regions supported by HMS product:
- Continental United States
- Alaska
- Hawaii
- Mexico
- Parts of Canada

HMS products are available in ASCII, graphical and Geographic Information System (GIS) formats.

**Description of HMS operational fire and products:**

 HMS's smoke analysis is based on visual classification of plumes using GOES-16 and GOES-17 ABI true-color imagery available during the sunlit part of the orbit.

 The HMS is an interactive environmental satellite image display and graphical interface system that was developed by NESDIS. Analysts manually compare automated fire detections to the mid-wave infrared (MWIR; 3.9 $\mu$m) images used to produce them to ensure each fire exists and detections deemed to be false are removed, and fires that are not automatically detected are added manually. Visible satellite imagery is also used by analysts to identify fires that may be too small to be automatically detected, either because they do not produce sufficient heat or because they are obscured by a tree canopy (Rolph et al., 2009). Land-cover data and power-plant locations compliment satellite imagery to help HMS analysts confirm whether automatic detections are fires (Ruminski et al., 2006). The HMS office makes a distinction between all detected fires (hereafter HMS hotspots) and fires an HMS analyst has confirmed to produce a substantial amount of smoke (hereafter HYSPLIT points).

HYSPLIT points are a subset of the HMS hotspots; they are fire detections where an analyst also visually confirms the presence of smoke using visible satellite imagery. Each HYSPLIT point is assigned a latitude, longitude, date, time, and duration. The number of HYSPLIT points analyzed, multiplied by their respective duration is proportional to visual smoke produced for the land cover in which it is burning, and the analyst determines this relationship. The relationship between smoke quantity (as defined by particulate matter concentration ($\mu$g m$^{−3}$) or kilograms of emissions for example) and HYSPLIT points is variable and does not account for land cover. Large wildfires are represented by dozens of HYSPLIT points spread over many square kilometers. The durations for these HYSPLIT points are estimated using the difference between the latest and earliest time for hotspots aggregated into a 20 km grid.

After identifying HYSPLIT points, HMS analysts use imagery from multiple NOAA and NASA satellites to identify the geographic extent of smoke plumes (Rolph et al., 2009; Ruminski et al., 2006). Smoke detection is done with visible-band imagery occasionally assisted by infrared to distinguish between clouds and smoke when possible (Ruminski et al., 2006). Given the limitations of the satellite data (mostly obscuration of smoke due to cloud cover), the number and extent of smoke plumes within this dataset represents a conservative estimate. Due to the limitations of visible satellite imagery and time constraints, no information about the vertical location or extent of smoke plumes is provided. In 2006 HMS analysts began providing estimates of smoke-plume concentrations (e.g., 5, 16, and 27 $\mu$g m$^{−3}$).



**Challenges and Uncertainties**

There is no published validation of the HMS smoke-plume extent data. It could have systematic biases because discrimination of smoke is easier over some areas, seasons, and under certain synoptic weather conditions. The largest uncertainty is at the edges of the plumes as the product is affected by the coarse spatial resolution of GOES imagery. The temporal information is also a source of uncertainty. During the day, the extent can be assessed hourly. However, this is not possible overnight. Smoke is sometimes transported to areas with anthropogenic haze pollution. In some cases the smoke will mix with and become indistinguishable from the anthropogenic haze pollution. The greater the distance traveled by a smoke plume, the more challenging it is to distinguish between smoke and anthropogenic haze. This challenge is particularly pronounced for aged smoke impacting the Southeast United States. HMS smoke polygons represent limited daytime snapshots of column smoke presence and do not contain information about the vertical location of smoke – i.e., whether the smoke is aloft or near the surface. HMS may be a poor indicator of surface smoke where smoke is expected to be mostly aloft, such as over states in the Midwest and Northeast that do not generate large amounts of smoke from wildfires and prescribed fires but instead receive smoke transported from other regions.

**Data Information:**

Satellite fire detection data are displayed using a color scheme based on the pixel's fire radiative power (FRP), given in MegaWatts [MW]. A fill value of -999 is used when no FRP retrieval is available.


**Note for users:**

The initial HMS product for the current day is created and updated by a satellite analyst roughly between 8 am and 10 am Eastern Time. After 10 am, the analysis is fine-tuned as time permits as additional satellite data becomes available. Areas of smoke are analyzed and added to the analysis during daylight hours as visible satellite imagery becomes available. The product is finalized and "completed" for the archive the following morning - generally by around 8 am.

# Import Required Libraries and Define HMS Class

```python
                                          

import os
import json
from datetime import datetime, date, timedelta

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

import wget
import geopandas as gpd
import fiona
from shapely.geometry import shape, Point, mapping, Polygon
from osgeo import gdal, ogr

import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
class HMS:
    """
    The HMS (Hazard Mapping System) class provides functionalities to access, analyze,
    and visualize smoke plume data from satellite observations. This includes downloading
    and processing Shapefiles, extracting specific information about smoke plumes, and
    plotting smoke plume geometries on a map.

    Required Modules:
        - wget
        - geopandas
        - cartopy
        - fiona
        - json
        - pandas
        - matplotlib
        - numpy
        - shapely
        - gdal

    Attributes:
        base_url (str): Base URL for the HMS smoke plume Shapefiles.
        current_date (datetime.date): Current date for data access.
        data_dir (str): Directory to store downloaded data.
        file_name (str): Full file path for the downloaded data file.
        hms_gpdf (GeoDataFrame): GeoDataFrame containing smoke plume data.

    Methods:
        __init__(self, base_url, data_dir): Initializes the HMS object with the base URL and data directory.
        set_date(self, dt): Sets the current date for data access.
        isvalid(self, geom): Checks if the provided geometry is valid.
        get_valid_polygons(self): Filters and returns valid smoke plume geometries as a GeoDataFrame.
        get_data(self, dt): Downloads and loads HMS Shapefile data for the specified date.
        get_satellite(self, id): Retrieves the satellite name associated with a smoke plume.
        get_start_date(self, id): Retrieves the start date of a smoke plume.
        get_end_date(self, id): Retrieves the end date of a smoke plume.
        get_density(self, id): Retrieves the smoke plume density.
        get_geometry(self, id): Retrieves the geometry of a smoke plume.
        get_duration(self, id): Calculates and returns the duration of a smoke plume.
        get_info(self, id): Retrieves all relevant information about a smoke plume.
        plot_plumes(self, pol, bounds): Plots smoke plumes on a Lambert Conformal map.
        find_intersects(self, poly): Finds smoke plumes intersecting with a given polygon.
        set_raster_grid(self, lon_west, lon_east, lat_north, lat_south, inc): Sets up the raster grid for smoke plume visualization.
        get_raster(self): Retrieves rasterized smoke plume data.
    """

    def __init__(self, base_url='https://satepsanone.nesdis.noaa.gov/pub/FIRE/web/HMS/Smoke_Polygons/Shapefile',
                 data_dir='hms_data'):
        """Initialize the HMS object with the base URL for data and a directory for storage."""
        self.base_url = base_url
        self.current_date = date.today()
        self.data_dir = data_dir
        if not os.path.exists(self.data_dir):
            os.makedirs(self.data_dir)

    def set_date(self, dt):
        """Set the current date for data access."""
        self.current_date = dt

    def isvalid(self, geom):
        """Check if the given geometry is valid."""
        try:
            shape(geom)
            return True
        except:
            from IPython import embed
            embed()
            return False

                                   
                                                                                        
                                                                                
                                        
                                                                           
                                        
                                                                
                                                           

    def get_data(self, dt):
        """Download and read the HMS Shapefile data for the specified date. Return True if successful."""
        self.current_date = dt
        self.file_name_prefix = f'hms_smoke{self.current_date.strftime("%Y%m%d")}'
        self.file_name = os.path.join(self.data_dir, self.file_name_prefix + '.zip')
        self.url = f"{self.base_url}/{self.current_date.year}/{self.current_date.month:02d}/{self.file_name_prefix}.zip"

        if not os.path.exists(self.file_name):
            try:
                print(f'Downloading {self.url} to {self.data_dir}')
                wget.download(self.url, out=self.file_name)
            except Exception as err:
                print(f'WARNING: wget error: {err} while accessing {self.url}')
                return False

        try:
            self.hms_gpdf = gpd.read_file(f'zip:///content/{self.file_name}')
        except Exception as err:
            print(f'WARNING: Geopandas read error: {err}')
            return False

        self.smoke_plume_count = len(self.hms_gpdf)
        self.crs = self.hms_gpdf.crs
        return True

    def get_satellite(self, id):
        """Retrieve the satellite name associated with a specific smoke plume."""
        return self.hms_gpdf.iat[id, 0]

    def get_start_date(self, id):
        """Retrieve the start date of a specific smoke plume."""
        return datetime.strptime(self.hms_gpdf.iat[id, 1], '%Y%j %H%M')

    def get_end_date(self, id):
        """Retrieve the end date of a specific smoke plume."""
        return datetime.strptime(self.hms_gpdf.iat[id, 2], '%Y%j %H%M')

    def get_density(self, id):
        """Retrieve the density of a specific smoke plume."""
        return self.hms_gpdf.iat[id, 3]

    def get_geometry(self, id):
        """Retrieve the geometry of a specific smoke plume."""
        return self.hms_gpdf.iat[id, 4]

    def get_duration(self, id):
        """Calculate and return the duration of a specific smoke plume."""
        start = datetime.strptime(self.hms_gpdf.iat[id, 1], '%Y%j %H%M')
        end = datetime.strptime(self.hms_gpdf.iat[id, 2], '%Y%j %H%M')
        return end - start

    def get_info(self, id):
        """Retrieve all relevant information about a specific smoke plume."""
        return (self.get_satellite(id), self.get_start_date(id),
                self.get_end_date(id), self.get_density(id), self.get_geometry(id))

    def plot_plumes(self, pol=None, bounds=[-122., -72., 10., 50.]):
        """Plot smoke plumes on a Lambert Conformal map within the specified bounds."""
        plt.figure(figsize=(5, 3))
        ax = plt.axes(projection=ccrs.LambertConformal())
        ax.set_extent(bounds, crs=ccrs.PlateCarree())
        ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.75)
        ax.add_feature(cfeature.STATES, linewidth=0.5)
        ax.add_geometries(self.hms_gpdf["geometry"], crs=ccrs.PlateCarree(),
                          facecolor='b', edgecolor='black', alpha=0.5)
        if pol is not None:
            ax.add_geometries([pol], crs=ccrs.PlateCarree(), facecolor='r', edgecolor='black', alpha=0.5)

                                            
                                           
                                                              
                                                   
                                    
                  

                                      
                                                                                                
                                                    
                                                    

    def find_intersects(self, poly):
        """Find and return indices of smoke plumes intersecting with the specified polygon."""
                                                    
                                                    
        intr_list = []
        for idx, geom in enumerate(self.hms_gpdf['geometry']):
          if geom.intersects(poly):
            intr_list.append(idx)
          else:
            continue
        return intr_list

    def set_raster_grid(self, lon_west=-125.0, lon_east=-65, lat_north=55.0, lat_south=10.0, inc=0.0083333333):
        """Set up the raster grid boundaries and resolution for smoke plume visualization."""
        self.grid_bounds = [lon_west, lat_south, lon_east, lat_north]
        self.grid_inc = inc
        nlon = int((lon_east - lon_west) / inc)
        nlat = int((lat_north - lat_south) / inc)
        lon = np.linspace(lon_west, lon_east, nlon)
        lat = np.linspace(lat_north, lat_south, nlat)
        self.lons, self.lats = np.meshgrid(lon, lat)

    def get_raster(self):
        """Retrieve rasterized smoke plume data using the specified grid bounds and increment."""
        try:
            os.system(f'unzip {self.file_name} -d {self.data_dir}')
        except Exception as err:
            print(f'WARNING: encountered error while unzipping: {err}')

        ds = gdal.Rasterize('tmpf', os.path.join(self.data_dir, self.file_name_prefix + '.shp'),
                            xRes=self.grid_inc, yRes=self.grid_inc, allTouched=True,
                            outputBounds=self.grid_bounds, burnValues=1,
                            outputType=gdal.GDT_Byte)

        mask = ds.ReadAsArray()
        return mask
```

# Write Data in Pandas Dataframe

```python
                                                            
dt         = date(2018,7,1)
end_date   = date(2018,10,30)
inc        = timedelta(days=1)


                                         

poly = Polygon([(-117.88984, 34.34471), (-117.76057, 34.34545),
                (-117.75701, 34.23030),(-117.92431, 34.26690),
                (-117.88984, 34.34471)])

df = pd.DataFrame([])                                                             

while dt < end_date:
  if rdr.get_data(dt):
      try:
          intr_arr = rdr.find_intersects(poly)
      except Exception as err:
          print(f'Error: encountered error while finding intersects: {err}')
          intr_arr = []
          dt += inc
          continue

                                                         
      n = 1 if len(intr_arr) > 0 else 0
      row = pd.DataFrame([{'Date': dt, 'smoke': n}])
      df = pd.concat([df, row])

                      
  dt += inc

df = df.set_index(pd.DatetimeIndex(df['Date']))
del df['Date']

                                                                       
df = df.resample("ME").sum()
df.plot()
```

# Applications

HMS data are an important source for tracking smoke plumes, however it does not provide vertical information on smoke plumes and thus does not reliably reflect the presence of smoke at ground level. Nevertheless, the HMS can be instrumental in identifying days when regional PM2.5 concentrations are likely influenced by smoke. Kaulfus et al. (2017) used HMS data to derive a climatology of smoke occurrence over the Continental United States (CONUS) region. The results indicated that smoke is most frequently found over the Great Plains. Using additional information the authors found that a substantial portion (20%) of the 24 h federal standard for particulate pollution exceedance events in the CONUS region occur when smoke is present. Figure 1 shows a spatial distribution of smoke events for the time period of August 1, 2005 through August 31, 2016 (133 months). Individual smoke polygons were merged creating a daily analysis of smoke areal extent. The daily smoke areal extent analysis was then utilized to the quantify number of days when smoke was present in the atmospheric column over a given location. In addition to this studies, other studies utilizing the HMS-detected smoke plumes have identified a strong positive association between the transported smoke plumes in the atmospheric column and collocated surface PM2.5 enhancement in the eastern US (Brey et al. 2018; Gunsch et al. 2018).

hms_smoke.png



Other studies have utilized the HMS data in conjuction with ground-based instrumentation to examine air quality under wild fire events. *O’Dell et al. (2023)* studied the yealy distribution of PM$_{2.5}$ concentration from the PurpleAir low-cost sensor network in wester US cities in 2020. The authors used the HMS data to identify smoke days and smoke-free days, and examined the relationship between indoor and outdoor particulate air quality. While the results on the influence of smoke on indoor air quality were limited by a lack of detail on indoor environment characteristics, as outdoor PM2.5 concentrations increased on smoke impacted days, indoor PM2.5 concentrations also increased, but the ratio of indoor to outdoor PM2.5 decreased.

smoke_vs_smoke_free.png

HMS data have also been used in a machine learning framework to examine wildfire impacts on O$_{3}$ in the Continental US between smoke and nonsmoke days. Lee et al. (2024) built a trainig dataset using the PM$_{2.5}$-criteria for each EPA site defined as the mean + 1 standard deviation (SD) of the daily surface PM$_{2.5}$ values on days with no HMS identified smoke (HMS = 0) from May to September for 2018–2023. Days with smoke were found to have an average of 11 μg m$^{–3}$ more PM$_{2.5}$ and 8 ppb higher maximum daily 8 h (MDA8) average O$_{3}$ concentrations than nonsmoke days. To estimate the smoke contributions to the O$_{3}$ MDA8, Generalized Additive Models were built for each site using the nonsmoke day data and up to 8 predictors. It was found that in 2023, the Central US experienced an unusually high number of exceedance days, 1522, with 52% of these impacted by smoke, while the Eastern US had fewer exceedance days, 288, with 78% of these impacted by smoke.
